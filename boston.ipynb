```python
import pandas as pd
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=fetch_california_housing()
df
```

```python
dataset=pd.DataFrame(df.data)
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

```python
# independetn featurs and dependent features
x=dataset
y=df.target
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.30, random_state=42)
```

```python
x_train
```

```python
# standardize the dataset 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
scaler.inverse_transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression
# cross validation 
from sklearn.model_selection import cross_val_score
```

```python
regression= LinearRegression()
regression.fit(x_train,y_train)
```

```python
mse=cross_val_score(regression,x_train,y_train,scoring="neg_mean_squared_error",cv=5)
```

```python
np.mean(mse)
```

```python
reg_pred=regression.predict(x_test)
```

```python
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred - y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(reg_pred,y_test)
```

```python
score
```

Ridge Regression Alogrithm

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

```

```python
ridge_regression=Ridge()
ridge_regression
```

```python
parameters={'alpha':[1,2,3,6,7,10,20,30,40,50,60,70,80,90]}
ridgecv=GridSearchCV(ridge_regression,parameters,scoring='neg_mean_squared_error',cv=5)
ridgecv.fit(x_train,y_train)
```

```python
print(ridgecv.best_params_)
```

```python
print(ridgecv.best_score_)
```

```python
ridgecv.predict(x_test)
```

```python
ridge_pred=ridgecv.predict(x_test)
```

```python
import seaborn as sns
sns.displot(ridge_pred - y_test, kind='kde')
```

```python
score=r2_score(ridge_pred,y_test)
```

```python
score
```

lasso regression 

```python
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso()
```

```python
parameters={'alpha':[0.1,0.5,1,5,10,20,50,100]}     
lassocv=GridSearchCV(lasso, parameters, scoring = 'neg_mean_squared_error', cv=5)
lassocv.fit(x_train,y_train)
```

```python
print(lassocv.best_params_) 
print(lassocv.best_score_)
```

```python
lasso_pred=lassocv.predict(x_test)
```

```python
import seaborn as sns
sns.displot(lasso_pred - y_test, kind='kde')
```
